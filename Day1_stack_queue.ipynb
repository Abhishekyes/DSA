### Stack Data Structure

A **stack** is a linear data structure that follows the Last In, First Out (LIFO) principle. This means that the last element added to the stack will be the first one to be removed. Think of it like a stack of plates; you add plates to the top and also remove plates from the top.

### Operations on a Stack

1. **Push**: Add an element to the top of the stack.
2. **Pop**: Remove and return the top element of the stack. If the stack is empty, this operation usually raises an error.
3. **Peek**: Return the top element of the stack without removing it. If the stack is empty, this operation usually raises an error.
4. **Is Empty**: Check if the stack is empty.
5. **Size**: Return the number of elements in the stack.

### Unique Characteristics

- **LIFO Order**: The most unique characteristic of a stack is its LIFO order of operations.
- **Single Access Point**: All operations (push, pop, peek) are performed at one end of the stack, known as the top.
- **Dynamic Size**: The size of the stack can grow or shrink dynamically as elements are added or removed.

Stacks are commonly used in various applications such as:
- **Function Call Management**: The call stack in programming languages keeps track of function calls.
- **Expression Evaluation**: Stacks are used to evaluate expressions, especially in compilers.
- **Undo Mechanisms**: Many applications use stacks to implement undo features.

```python
class Stack:
    def __init__(self):
        self.stack = []
        self.size = 0

    def push(self, item):
        self.stack = self.stack[:self.size] + [item]
        self.size += 1

    def pop(self):
        if self.is_empty():
            raise IndexError("pop from empty stack")
        item = self.stack[self.size - 1]
        self.stack = self.stack[:self.size - 1]
        self.size -= 1
        return item

    def peek(self):
        if self.is_empty():
            raise IndexError("peek from empty stack")
        return self.stack[self.size - 1]

    def is_empty(self):
        return self.size == 0

    def display(self):
        return self.stack[:self.size]

# Example usage:
stack = Stack()
stack.push(1)
stack.push(2)
stack.push(3)
print("Stack after pushes:", stack.display())
print("Peek:", stack.peek())
print("Pop:", stack.pop())
print("Stack after pop:", stack.display())
```

### Queue Data Structure

A **queue** is a linear data structure that follows the First In, First Out (FIFO) principle. This means that the first element added to the queue will be the first one to be removed. Think of it like a line of people waiting for a bus; the first person in line is the first one to get on the bus.

### Operations on a Queue

1. **Enqueue**: Add an element to the end of the queue.
2. **Dequeue**: Remove and return the front element of the queue. If the queue is empty, this operation usually raises an error.
3. **Front**: Return the front element of the queue without removing it. If the queue is empty, this operation usually raises an error.
4. **Is Empty**: Check if the queue is empty.
5. **Size**: Return the number of elements in the queue.

### Unique Characteristics

- **FIFO Order**: The most unique characteristic of a queue is its FIFO order of operations.
- **Two Access Points**: Elements are added at the rear (enqueue) and removed from the front (dequeue).
- **Dynamic Size**: The size of the queue can grow or shrink dynamically as elements are added or removed.

Queues are commonly used in various applications such as:
- **Task Scheduling**: Queues are used to manage tasks in operating systems and other applications.
- **Breadth-First Search**: Queues are used in graph traversal algorithms like BFS.
- **Print Spooling**: Queues are used to manage print jobs in a printer.

```python
class Queue:
    def __init__(self):
        self.queue = []
        self.front_index = 0
        self.rear_index = -1
        self.size = 0

    def enqueue(self, item):
        self.queue.append(item)
        self.rear_index += 1
        self.size += 1

    def dequeue(self):
        if self.is_empty():
            raise IndexError("dequeue from empty queue")
        item = self.queue[self.front_index]
        self.front_index += 1
        self.size -= 1
        return item

    def front(self):
        if self.is_empty():
            raise IndexError("front from empty queue")
        return self.queue[self.front_index]

    def is_empty(self):
        return self.size == 0

    def display(self):
        return self.queue[self.front_index:self.rear_index + 1]

# Example usage:
queue = Queue()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
print("Queue after enqueues:", queue.display())
print("Front:", queue.front())
print("Dequeue:", queue.dequeue())
print("Queue after dequeue:", queue.display())
```

```python
import matplotlib.pyplot as plt

class Queue:
    def __init__(self):
        self.queue = []
        self.front_index = 0
        self.rear_index = -1
        self.size = 0

    def enqueue(self, item):
        self.queue.append(item)
        self.rear_index += 1
        self.size += 1
        self.visualize("Enqueue")

    def dequeue(self):
        if self.is_empty():
            raise IndexError("dequeue from empty queue")
        item = self.queue[self.front_index]
        self.front_index += 1
        self.size -= 1
        self.visualize("Dequeue")
        return item

    def front(self):
        if self.is_empty():
            raise IndexError("front from empty queue")
        return self.queue[self.front_index]

    def is_empty(self):
        return self.size == 0

    def display(self):
        return self.queue[self.front_index:self.rear_index + 1]

    def visualize(self, operation):
        plt.figure(figsize=(10, 2))
        plt.title(f"Queue after {operation}")
        if self.is_empty():
            plt.text(0.5, 0.5, 'Queue is empty', horizontalalignment='center', verticalalignment='center', fontsize=12)
        else:
            plt.bar(range(self.front_index, self.rear_index + 1), self.queue[self.front_index:self.rear_index + 1])
            plt.xlabel("Index")
            plt.ylabel("Value")
        plt.show()

# Example usage:
queue = Queue()
queue.visualize("Initialization")
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
print("Queue after enqueues:", queue.display())
print("Front:", queue.front())
print("Dequeue:", queue.dequeue())
queue.visualize("After Dequeue")
print("Queue after dequeue:", queue.display())
```
